```python
# Montando DataSet Faixa Etária

localidade = 503 # Cidade de São Paulo
# Lista de variáveis referentes a informações de Densidade Demográfica
vars_faixa_etaria= ['72','73','108', '109','111', '112','78', '79','115', '116','118','119', '121', '122', '124',  '125','127',  '128','130',  '131',  '133',  '134',  '136', '137', '139',  '140','142',  '143', '145', '146',  '148',  '149']
ano = 2000 # primeiro ano a ser considerado na base população Total

base_faixa_etaria  = []
max_ano = 2019
while ano<= max_ano:
    for i in vars_faixa_etaria:
        url = 'http://api-imp.seade.gov.br/v1/dados/' + str(localidade) + '/' + str(i) + '/'  + str(ano)
        try:
            response = requests.get(url)
            base_faixa_etaria.append({'ano':ano, 'variavel': response.json()['dados'][0]['variavel'], 'Cod_variavel':response.json()['dados'][0]['codigo_variavel'], 'Qtde': int(response.json()['dados'][0]['ano'][str(ano)].replace('.',''))})
        except:
            print( 'ano com falha: '+ str(ano))
    ano+=1
```

```python
base_faixa_etaria_ = pd.DataFrame(base_faixa_etaria) # montando DataFrame
base_faixa_etaria_.to_sql('base_faixa_etaria', db, schema='dbo', index=False, if_exists='replace')
base_faixa_etaria_
```
