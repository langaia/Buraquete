## Overview

Parte fundamental da atuação de um profissional de dados é a “construção de pontes”, isto é, encontrar nexos e conexões entre os dados e melhores decisões. É esta capacidade analítica, tantas vezes ressaltada em nossos encontros, que trabalharemos, majoritariamente, em nosso Projeto Integrador. 

Vocês trabalharão em grupos com a missão de desenvolver, nos próximos meses, um projeto que parta de um problema a ser resolvido, exploração e análise de dados que tragam mais informações sobre o cenário e que embasem, obviamente, possíveis soluções/intervenções.

A apresentação será na última aula do nosso cronograma, quando cada grupo terá 15min para apresentar seu trabalho. Nas próximas páginas vocês terão detalhes e instruções sobre como proceder

Em 2018, alguns grupos escolheram temas como “Água” e “Saúde”, com a intenção de desenvolverem soluções que gerassem impacto social, e iniciaram o planejamento a partir de um diagnóstico mais amplo sobre o cenário atual. ## Fases do Projeto

### Diagnóstico

Este é o momento em que mergulharemos de cabeça no problema para entender/aprender ao máximo sobre o contexto. Precisamos responder as seguintes perguntas:

- Qual a situação atual?
- Qual o histórico? Como chegamos até a situação atual?
- Como o problema escolhido por vcs evoluiu ao longo do tempo?
- Quais os principais aspectos que devemos “atacar” para ajudar na melhoria do quadro geral?

### Mapeamento

Uma vez que temos uma visão mais clara sobre o cenário e suas particularidades, é hora de determinar quais as fontes de dados que poderemos utilizar. Para ajudar no planejamento, podemos usar as seguintes perguntas como ponto de partida:

- Quais os dados que geralmente são utilizados?
- Existem outras fontes de dados que poderiam ajudar a ampliar a compreensão acerca do problema?
- Pensando de maneira criativa: quais outras fontes de dados (ex: dados abertos, dados internacionais, dados de mídias sociais etc), poderiam nos ajudar?

### Definindo as Perguntas

Uma vez que entendemos melhor as circunstâncias e contextos que orbitam e impactam o problema escolhido pelo seu grupo, é importante definir algumas perguntas centrais que vão nos ajudar a nortear o desenvolvimento de uma proposta de solução. Trabalhem na elaboração de 2 a 4 perguntas que, uma vez respondidas, trarão resultados tangíveis e mensuráveis.

### Elaborando Indicadores

Agora que já temos as fontes de dados e as perguntas centrais que precisamos responder para produzir um resultado prático, é o momento de definir um ou mais indicadores que nos auxiliarão no processo de mensuração dos efeitos da nossa proposta de intervenção.

### DataViz e Storytelling

O último passo é estruturar o projeto para ser apresentado. Além do roteiro em si, isto é, da maneira que a história será contada, é imprescindível que vocês elaborem ao menos 3 opções diferentes de visualização dos dados.

### Colocando em Prática

Cada grupo deverá planejar uma intervenção prática, de qualquer natureza, que esteja solidamente embasada por todo o trabalho analítico realizado nas fases anteriores.

As possibilidades são diversas. Pesquisem referências, cases e exemplos em diferentes segmentos.

E, o mais importante, SEJAM CRIATIVOS(AS)!!

O Portal Brasileiro de Dados Abertos é a ferramenta disponibilizada pelo governo para que todos possam encontrar e utilizar os dados e as informações públicas. O portal preza pela simplicidade e organização para que você possa encontrar facilmente os dados e informações que precisa. O portal também tem o objetivo de promover a interlocução entre atores da sociedade e com o governo para pensar a melhor utilização dos dados, promovendo impactos positivos sob os pontos de vista social e econômico.
Quais dados estão disponíveis aqui?

O portal tem o objetivo de disponibilizar dados relativos às mais variadas temáticas da administração pública. Por exemplo, dados da saúde suplementar, do sistema de transporte, de segurança pública, indicadores de educação, gastos governamentais, processo eleitoral, etc.

O portal contém apenas dados abertos. Isso significa que dados que contenham alguma restrição de acesso, tais como aquelas advindas de sigilo ou privacidade, estão fora do escopo deste portal.

O portal funciona como um catálogo federado que facilita a busca e uso de dados publicados pelos órgãos do governo. Dadas as limitações de recursos, os diversos órgãos e entidades da administração pública disponibilizam os dados conforme os cronogramas estabelecidos em seus Planos de Dados Abertos – PDA. Se os dados que você procura não estão no portal, verifique se o PDA do órgão responsável tem esses dados em seu cronograma de disponibilização futura. Caso não esteja previsto no cronograma, ainda é possível abrir uma solicitação de acesso à informação, com base no Capítulo IV do Decreto 8.777/2016.

## PNAD - Pesquisa Nacional por Amostra de Domicílios - PNAD

<br>
<img src="img/pnad.png" width="800" />
<br>


A Pesquisa Nacional por Amostra de Domicílios - PNAD, de periodicidade anual, foi encerrada em 2016, com a divulgação das informações referentes a 2015. Planejada para produzir resultados para Brasil, Grandes Regiões, Unidades da Federação e nove Regiões Metropolitanas (Belém, Fortaleza, Recife, Salvador, Belo Horizonte, Rio de Janeiro, São Paulo, Curitiba e Porto Alegre), ela pesquisava, de forma permanente, características gerais da população, educação, trabalho, rendimento e habitação, e, com periodicidade variável, outros temas, de acordo com as necessidades de informação para o País, tendo como unidade de investigação o domicílio. A PNAD foi substituída, com metodologia atualizada, pela Pesquisa Nacional por Amostra de Domicílios Contínua - PNAD Contínua,  que propicia uma cobertura territorial mais abrangente e disponibiliza informações conjunturais trimestrais sobre a força de trabalho em âmbito nacional.  

A PNAD foi iniciada no segundo trimestre de 1967, sendo os seus resultados apresentados com periodicidade trimestral até o primeiro trimestre de 1970. A partir de 1971, os levantamentos passaram a ser anuais, com realização no último trimestre, sendo interrompidos nos anos de realização dos Censos Demográficos. Em 1974-1975, foi levada a efeito uma pesquisa especial, denominada Estudo Nacional da Despesa Familiar - ENDEF e, durante a sua realização, o levantamento básico da PNAD foi interrompido. Em 1994, por razões excepcionais, a pesquisa não foi realizada. Ao longo do tempo, a PNAD passou por atualizações metodológicas, algumas restritas ao plano amostral, e outras relacionadas à abrangência e às conceituações dos aspectos pesquisados, em consonância com as recomendações internacionais.  

No decorrer dos 49 anos de sua existência, a PNAD constituiu um importante instrumento para formulação, validação e avaliação de políticas orientadas para o desenvolvimento socioeconômico da população e a melhoria das condições de vida no País.

Para acesso [clique aqui](https://www.ibge.gov.br/estatisticas/sociais/trabalho/9127-pesquisa-nacional-por-amostra-de-domicilios.html?=&t=o-que-e)

## Portal de Dados Abertos da Prefeitura de São Paulo

<br>
<img src="img/pmsp.png" width="800" />
<br>

O Portal de Dados Abertos da Prefeitura de São Paulo tem sua origem no Catálogo Municipal de Bases de Dados (CMBD). Previsto pelo Decreto Municipal nº 54.779/2014, o CMBD tem por finalidade divulgar uma listagem de todas as bases de dados produzidas pelos órgãos e entidades da administração municipal.

Organizado pela Coordenadoria de Promoção da Integridade da Controladoria Geral do Município (CGM-SP) e pela Secretaria Municipal de Desenvolvimento Urbano (SMDU), o Catálogo apresentou aos munícipes uma relação com todos os dados disponíveis produzidos pela Prefeitura. Ao todo, foram catalogadas 591 bases de informação, incluindo dicionários de dados e metadados. Mesmo que a maior parte das bases, propriamente, não estivessem disponíveis, o Catálogo tornou possível verificar os dados disponíveis e os órgãos detentores das informações, que puderam ser solicitadas via e-SIC.

A partir do Catálogo, a Prefeitura de São Paulo passou a trabalhar em uma plataforma livre que reunisse suas bases de dados e descrições, permitindo fácil acesso e pesquisa por parte de programadores, jornalistas, pesquisadores e cidadãos em geral. Chegou-se ao CKAN (The Comprehensive Kerbal Archive Network), ferramenta já adotada nas maiores cidades do mundo, para a construção deste Portal de Dados Abertos, que conferirá ainda mais transparência aos dados municipais.

O CKAN é livre, gratuito e específico para o desenvolvimento de catálogos de dados. Utilizado pelos governos dos Estados Unidos, Reino Unido, Alemanha, Japão, Brasil e outros 139 governos do mundo, entre administrações nacionais e locais, ele foi customizado para a Prefeitura de São Paulo pela Prodam, a Empresa de Processamento de Dados do Município, a partir da evolução dos trabalhos desenvolvidos pela Controladoria Geral do Município e pela Secretaria Municipal de Desenvolvimento Urbano.

Por meio do CKAN, cada unidade da administração municipal poderá inserir suas bases e manter um fluxo de atualização constante. Gradativamente, todas as bases municipais passarão a figurar neste Portal de Dados Abertos.

Para aprimorar a ferramenta, a CGM e a SMDU também farão processos colaborativos voltados a desenvolvedores e cidadãos de maneira geral. As atividades podem ser acompanhadas no projeto Café Hacker. ## Compara Brasil

<br>
<img src="img/compara_brasil.png" width="800" />

O Compara Brasil é um portal de livre acesso com dados sobre as finanças públicas dos três níveis de governo. Seu objetivo é disponibilizar para a sociedade brasileira uma variedade de consultas e pesquisas aos dados fiscais dos municípios, dos estados e da União de forma simples e rápida. Através dos sistemas que compõem o Compara Brasil, o usuário monta planilhas, gráficos, rankings e mapas definindo seus próprios parâmetros.

Compara Brasil cumpre uma função social fundamental, a de tornar simples e acessível, em um único site, a maior parte dos dados disponíveis sobre as finanças públicas no Brasil. Baseado em fontes oficiais, o portal se pauta pelo princípio da transparência com acessibilidade.

O Portal Compara Brasil é uma realização da Frente Nacional de Prefeitos (FNP), desenvolvido em parceria com a empresa Aequus Consultoria.
FNP - Frente Nacional de Prefeitos

Com 24 anos de história, a Frente Nacional de Prefeitos é composta atualmente por 250 municípios filiados. O número inclui a participação de 26 capitais brasileiras e de mais de cem cidades com população acima de 100 mil habitantes, agrupando, aproximadamente, 40% de toda a população brasileira.

A FNP é a única representante dos municípios brasileiros a ser dirigida exclusivamente por prefeitos e prefeitas em efetivo exercício de mandato. Suprapartidária, sua missão é resgatar o princípio da autonomia municipal, garantido na Constituição Federal. Para isso, sempre que necessário, defende a redefinição do Pacto Federativo e propõe a participação ativa dos entes locais nas questões urbanas, além da interlocução ampla e democrática com os governos estaduais e federal e a sociedade civil organizada.

As ações da FNP são coordenadas por uma diretoria própria, eleita a cada biênio pelos filiados da entidade. A diretoria é composta pelo presidente; os vice-presidentes internacional, nacionais e regionais; o secretário-geral e os vice-presidentes temáticos, que envolvem 24 áreas de atuação, como saúde, educação, desenvolvimento econômico local. A entidade possui ainda um conselho fiscal.

A agenda municipalista da entidade é construída de maneira participativa, envolvendo todos os municípios filiados. A formulação feita no dia a dia ganha ênfase nos encontros nacionais promovidos pela FNP, que, pelo menos uma vez ao ano, organiza a Reunião Geral, evento de instância máxima para as deliberações da entidade. Com a participação de prefeitos, gestores municipais, parlamentares, representantes dos governos e especialistas em diversos assuntos, as reuniões gerais debatem os principais temas das cidades.
Aequus Consultoria

A Aequus Consultoria iniciou suas atividades em 1995, com foco de atuação nas finanças públicas municipais, sendo seus mais importantes produtos os anuários sobre as finanças públicas municipais para os estados do Espírito Santo, Minas Gerais e Rio de Janeiro, além de São Paulo e Multi Cidades - Finanças dos Municípios do Brasil que, atualmente, são desenvolvidos em parceria com a Frente Nacional de Prefeitos (FNP). A empresa elabora ainda outras publicações sobre as receitas e despesas públicas para cidades específicas.

Os anuários da Aequus Consultoria são publicações com informações analisadas sobre a evolução dos principais itens da receita e da despesa dos municípios, contendo diversos indicadores, rankings, gráficos, além de artigos sobre temas atuais da administração pública.

As publicações são distribuídas gratuitamente e direcionadas aos prefeitos, secretários municipais de finanças, vereadores, deputados estaduais e federais, senadores, órgãos estaduais, universidades, instituições públicas federais, institutos de pesquisa, instituições financeiras, meios de comunicação e uma ampla gama de empresas privadas.

A Aequus Consultoria também realiza estudos, pesquisas, consultorias e sistemas voltados para o campo do gerenciamento das finanças públicas.
