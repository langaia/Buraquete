```python
import pandas as pd 
import sqlite3  
```

```python
df_tarifa = pd.read_csv(r'C:\Users\Usuario\Desktop\DigitalHouse_DataScience\AulasPresenciais\Projeto_integrador\tarifa.txt',delimiter=';', decimal = ',')
```

```python
df_tarifa['VIGENCIA'] = df_tarifa['VIGENCIA'].astype('str')
```

```python
df_tarifa['ANO'] = df_tarifa.VIGENCIA.str.slice(6, 10).astype(int)
df_tarifa['MES'] = df_tarifa.VIGENCIA.str.slice(3, 5).astype(int)
```

```python
df_tarifa = df_tarifa.sort_values(by=['ANO', 'MES'])
df_tarifa = df_tarifa.reset_index(drop=True)
```

```python
df_tarifa_cp = []
for i in range(0,df_tarifa.shape[0]):
    ano = df_tarifa['ANO'][i]
    mes = df_tarifa['MES'][i]
    valor = df_tarifa['VALOR'][i]
    anomes= ano*100 + mes
    try: 
        anomes_n = df_tarifa['ANO'][i+1]*100 + df_tarifa['MES'][i+1]
    except: 
        anomes_n = anomes
        df_tarifa_cp.append({'ANO':ano,'MES':mes,'VALOR':valor})

    if anomes_n == anomes: 
         pass
    else:
        df_tarifa_cp.append({'ANO':ano,'MES':mes,'VALOR':valor})

    while anomes < anomes_n:
        if mes==12:
            mes=1
            ano+=1
        else:
            mes+=1
        if anomes_n == ano*100 + mes: 
            pass
        else:
            df_tarifa_cp.append({'ANO':ano,'MES':mes,'VALOR':valor})
        anomes= ano*100 + mes
    
```
