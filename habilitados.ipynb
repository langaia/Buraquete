```python
#Importar bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv('../Integrador/Fontes de Dados/RENACH/habilitados_sexo_categoria_idade.csv', encoding='utf8')
```

```python
#Remove a sujeira da categoria
df['Categoria']=df.Categoria.str.slice(0,2)
```

```python
#Condutores tipo A
a=[]
for i in range(0,df.shape[0]):
    categoria = df.Categoria[i]
    if "A" in categoria:
        a.append(bool(True))
    else:
        a.append(bool(False))

#Condutores tipo B
b=[]
for i in range(0,len(df.Categoria)):
    categoria = df.Categoria[i]
    if "B" in categoria:
        b.append(bool(True))
    else:
        b.append(bool(False))   
        
#Condutores tipo C
c=[]
for i in range(0,len(df.Categoria)):
    categoria = df.Categoria[i]
    if "C" in categoria:
        c.append(bool(True))
    else:
        c.append(bool(False))
        
#Condutores tipo D
d=[]
for i in range(0,len(df.Categoria)):
    categoria = df.Categoria[i]
    if "D" in categoria:
        d.append(bool(True))
    else:
        d.append(bool(False))
        
#Condutores tipo E
e=[]
for i in range(0,len(df.Categoria)):
    categoria = df.Categoria[i]
    if "E" in categoria:
        e.append(bool(True))
    else:
        e.append(bool(False))
```

```python
#Cria as colunas

df['T-A'] =a
df['T-B'] =b
df['T-C'] =c
df['T-D'] =d
df['T-E'] =e
```

```python
#Cria coluna Carros
df['Carros'] = df['T-B'] + df['T-C'] + df['T-D'] + df['T-E']
```
